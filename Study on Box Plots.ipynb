## Aim of Investigation

### Summarise the history of the box plot and situations in which its used 
In, 1969, mathematician John W. Tukey introduced the box plot which over time has seen other variations being stemmed from it. 
[https://en.wikipedia.org/wiki/Box_plot] 
<br>
![Tukey](Tukey1.gif)
<br>
As viewed in the above image, Tukey's original definition of a box plot contained hinges instead of quartiles (H_1 & H_3). 
Other differentiations included;
* outliers are indicated separately
* whiskers are dashed, ending with dashed crossbars at "adjacent values" 
* extended the whiskers all the way to the extreme data points
* lacked horizontal crossbars
* an unfilled dot at the maximum and a hatched horizontal strip at the minimum

<br>

**What is a BoxPlot?**<br>
*A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.*   https://www150.statcan.gc.ca/n1/edu/power-pouvoir/ch12/5214889-eng.htm<br> 
The Box Plot is a popular and simple way of representing statistical data.
Box Plots are described as non-parametric, meaning that they shows a true, non-biased view of the distribution of data without making any assumptions regarding the population distribution.
Boxplots can be drawn horizontally or vertically with their name coming from the box in the middle
<br>
*Box and whisker plots are ideal for comparing distributions because the centre, spread and overall range are immediately apparent.* https://www150.statcan.gc.ca/n1/edu/power-pouvoir/ch12/5214889-eng.htm

<br>
Boxplots are a popular choice to be used when obvserving large volumes of data and also when there are more than two data sets to compare. <br>
They are useful to determine if a distribtion is skewed and also if there are any unusual observations i.e outliers, without being influenced.
<br><br>

![Boxplot](boxplot.gif)

<br>
In a Box Plot: 

* the ends of the box are the upper and lower quartiles, so the box spans the interquartile range
* the median is marked by a vertical line inside the box
* the whiskers are the two lines outside the box that extend to the highest and lowest observations. 
[https://www150.statcan.gc.ca/n1/edu/power-pouvoir/ch12/5214889-eng.htm]


<br>

Boxplots are used in explainatory data analysis 
*The box plot (a.k.a. box and whisker diagram) is a standardized way of displaying the distribution of data based on the five number summary: minimum, first quartile, median, third quartile, and maximum.* [http://www.physics.csbsju.edu/stats/box2.html]
[https://chartio.com/resources/tutorials/what-is-a-box-plot/]

### Explain any relevant terminology such as terms quartile and percentile

![Boxplot](boxplot.gif)

#### Upper Extreme (Maximum) / Lower Extreme (Minimum)
The extremes of a dataset are the highest and lowest values, and the endpoints of the whiskers on a box plot. When a set of data is ordered, the first and last values are easily identified as the  the extremes in the given set. [https://www.youtube.com/watch?v=FXTXs6OMp-4]
[https://www.ck12.org/statistics/box-and-whisker-plots/lesson/Box-and-Whisker-Plots-MSM6/]

<br>

#### Median
The median refers to the central point of the set of data. This does not necessarily mean it will be exactly in the centre of the boxplot itself when plotted out. Medians can be found throughout box plots because of how each set of data is distributed. When the median of a set it marked, it  divides the dataset into two halves, giving you the lower and upper quartiles. [https://socratic.org/questions/is-the-median-always-in-the-exact-middle-of-a-boxplot]

<br>

#### Lower Quartile (25th Percentile)
Once the median is identified, this then divides the dataset into two halves.
The lower quartile is determined by finding the centre (median) in the lower half, i.e. The lower quartile ( Q 1 ) is the median of the lower half.
[http://www.flinders.edu.au/slc_files/Documents/Red%20Guides/Box%20and%20Whisker%20Plots.pdf]
[https://www.varsitytutors.com/hotmath/hotmath_help/topics/box-and-whisker-plots]

<br>

#### Upper Quartile (75th Percentile)
The same process as determining the lower quartile is used for calculating the Upper Quartile. The median of the higher half of data is the Upper Quartile of the overall set of data.
<br>
![Percentile](percentile.png)
<br>

#### Interquartile Range
The interquartile range is the difference between the upper and lower quartiles. 
*The IQR is a very useful measurement. It is useful because it is less influenced by extreme values as it limits the range to the middle 50% of the values.*
[http://www.flinders.edu.au/slc_files/Documents/Red%20Guides/Box%20and%20Whisker%20Plots.pdf]
<br>
It is worth noting that another term for the Quartile is the Percentile. 

<br>

#### Whisker
The whiskers of a box plot are the lines that extend out from the box plot on a graph. The whiskers extend out to connect to the min and max values of the dataset. <br>
The "whiskers" show how large the "spread" is of the data. It is rare that a boxplot is graphed without any whiskers.
[https://www.purplemath.com/modules/boxwhisk.htm]
<br>

#### Outlier / single data point
Outliers are the data identified as being beyond or drastically outside the other contents of a dataset. These can skew the overall results of a dataset explaining why sometimes outliers are discarded or excluded to allow for a more accurate representation of the set in the data output. 
*The spacings between the different parts of the box indicate the degree of dispersion (spread) and skewness in the data, and show outliers.*
[https://en.wikipedia.org/wiki/Box_plot]
*An outlier can also be stated as a value that lies outside the overall pattern of a distribution and thus can affect the overall data series.* [https://www.whatissixsigma.net/box-plot-diagram-to-identify-outliers/]

Box plots are a help method of identifying outliers. Any data lying beyond the upper and lower limits of a data set are considered outliers.
<br>
*The very purpose of this diagram is to identify outliers and discard it from the data series before making any further observation so that the conclusion made from the study gives more accurate results not influenced by any extremes or abnormal values.* [https://www.whatissixsigma.net/box-plot-diagram-to-identify-outliers/]
