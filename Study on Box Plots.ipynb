### Summarise the history of the box plot and situations in which its used 
In, 1969, mathematician John W. Tukey introduced the box plot which over time has seen other variations being stemmed from it. 
[https://en.wikipedia.org/wiki/Box_plot] 
<br>
![Tukey](Tukey1.gif)
<br>
As viewed in the above image, Tukey's original definition of a box plot contained hinges instead of quartiles (H_1 & H_3). 
Other differentiations included;
* outliers are indicated separately
* whiskers are dashed, ending with dashed crossbars at "adjacent values" 
* extended the whiskers all the way to the extreme data points
* lacked horizontal crossbars
* an unfilled dot at the maximum and a hatched horizontal strip at the minimum

<br>

**What is a BoxPlot?**<br>
*A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.*   https://www150.statcan.gc.ca/n1/edu/power-pouvoir/ch12/5214889-eng.htm<br> 
The Box Plot is a popular and simple way of representing statistical data.
Box Plots are described as non-parametric, meaning that they shows a true, non-biased view of the distribution of data without making any assumptions regarding the population distribution.
Boxplots can be drawn horizontally or vertically with their name coming from the box in the middle
<br>
*Box and whisker plots are ideal for comparing distributions because the centre, spread and overall range are immediately apparent.* https://www150.statcan.gc.ca/n1/edu/power-pouvoir/ch12/5214889-eng.htm

<br>

![Boxplot](boxplot.gif)

<br>
In a Box Plot: 

* the ends of the box are the upper and lower quartiles, so the box spans the interquartile range
* the median is marked by a vertical line inside the box
* the whiskers are the two lines outside the box that extend to the highest and lowest observations. 
[https://www150.statcan.gc.ca/n1/edu/power-pouvoir/ch12/5214889-eng.htm]


<br>

Boxplots are used in explainatory data analysis 
*The box plot (a.k.a. box and whisker diagram) is a standardized way of displaying the distribution of data based on the five number summary: minimum, first quartile, median, third quartile, and maximum.* [http://www.physics.csbsju.edu/stats/box2.html]
[https://chartio.com/resources/tutorials/what-is-a-box-plot/]
<br>
Boxplots are a popular choice to be used when obvserving large volumes of data and also when there are more than two data sets to compare. They can display data from multiple sources and display results in a single graph.
They are useful to determine if a distribtion is skewed and also if there are any unusual observations i.e outliers, without being influenced.
*This type of graph is used to show the shape of the distribution, its central value, and its variability.*
[https://www.statcan.gc.ca/edu/power-pouvoir/ch12/5214889-eng.htm]

### Demonstrate the use of the box plot using data of your choosing


```python
import numpy as np
import matplotlib.pyplot as plt

# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.boxplot.html
# https://matplotlib.org/gallery/pyplots/boxplot_demo_pyplot.html#sphx-glr-gallery-pyplots-boxplot-demo-pyplot-py

np.random.seed(1234567)

spread = np.random.rand(20) * 60
center = np.ones(5) * 80
flier_high = np.random.rand(10) * 60 + 80
flier_low = np.random.rand(10) * -50
data = np.concatenate((spread, center, flier_high, flier_low))

fig1, ax1 = plt.subplots()
ax1.set_title('Basic Plot')
ax1.boxplot(data)

plt.show
```

### Explain any relevant terminology such as terms quartile and percentile

![Boxplot](boxplot.gif)

#### Upper Extreme (Maximum) / Lower Extreme (Minimum)
The extremes of a dataset are the highest and lowest values, and the endpoints of the whiskers on a box plot. When a set of data is ordered, the first and last values are easily identified as the  the extremes in the given set. [https://www.youtube.com/watch?v=FXTXs6OMp-4]
[https://www.ck12.org/statistics/box-and-whisker-plots/lesson/Box-and-Whisker-Plots-MSM6/]

<br>

#### Median
The median refers to the central point of the set of data. This does not necessarily mean it will be exactly in the centre of the boxplot itself when plotted out. Medians can be found throughout box plots because of how each set of data is distributed. When the median of a set it marked, it  divides the dataset into two halves, giving you the lower and upper quartiles. [https://socratic.org/questions/is-the-median-always-in-the-exact-middle-of-a-boxplot]

<br>

#### Lower Quartile (25th Percentile)
Once the median is identified, this then divides the dataset into two halves.
The lower quartile is determined by finding the centre (median) in the lower half, i.e. The lower quartile ( Q 1 ) is the median of the lower half.
[http://www.flinders.edu.au/slc_files/Documents/Red%20Guides/Box%20and%20Whisker%20Plots.pdf]
[https://www.varsitytutors.com/hotmath/hotmath_help/topics/box-and-whisker-plots]

<br>

#### Upper Quartile (75th Percentile)
The same process as determining the lower quartile is used for calculating the Upper Quartile. The median of the higher half of data is the Upper Quartile of the overall set of data.

<br>

#### Interquartile Range
The interquartile range is the difference between the upper and lower quartiles. 
*The IQR is a very useful measurement. It is useful because it is less influenced by extreme values as it limits the range to the middle 50% of the values.*
[http://www.flinders.edu.au/slc_files/Documents/Red%20Guides/Box%20and%20Whisker%20Plots.pdf]
<br>
It is worth noting that another term for the Quartile is the Percentile. 

<br>

#### Whisker
The whiskers of a box plot are the lines that extend out from the box plot on a graph. The whiskers extend out to connect to the min and max values of the dataset. <br>
The "whiskers" show how large the "spread" is of the data. It is rare that a boxplot is graphed without any whiskers.
[https://www.purplemath.com/modules/boxwhisk.htm]
<br>

#### Outlier / single data point
Outliers are the data identified as being beyond or drastically outside the other contents of a dataset. These can skew the overall results of a dataset explaining why sometimes outliers are discarded or excluded to allow for a more accurate representation of the set in the data output. 
*The spacings between the different parts of the box indicate the degree of dispersion (spread) and skewness in the data, and show outliers.*
[https://en.wikipedia.org/wiki/Box_plot]
*An outlier can also be stated as a value that lies outside the overall pattern of a distribution and thus can affect the overall data series.* [https://www.whatissixsigma.net/box-plot-diagram-to-identify-outliers/]

Box plots are a help method of identifying outliers. Any data lying beyond the upper and lower limits of a data set are considered outliers.
<br>
*The very purpose of this diagram is to identify outliers and discard it from the data series before making any further observation so that the conclusion made from the study gives more accurate results not influenced by any extremes or abnormal values.* [https://www.whatissixsigma.net/box-plot-diagram-to-identify-outliers/]

### Compare the box plot to alternatives
<br>

#### Histograms
*A histogram is a type of bar chart that graphically displays the frequencies of a data set. ... The only difference between a histogram and a bar chart is that a histogram displays frequencies for a group of data, rather than an individual data point; therefore, no spaces are present between the bars.*
<br>
A box plot shows precise values (e.g. mean) which the histogram does not display.<br>

However where a box plot falls down in this comparison is when it comes to data containing recurring values / groups of numbers. If this was graphed on a box plot, the values would average out allowing the distribution to look normal. Whereas with a histogram, it displays the peaks in data allowing to observe and compare the results clearly.
<br>
![Histogram](histo.PNG)
![V.Histogram](histo2.PNG)

[https://www.brighthubpm.com/six-sigma/58254-box-plots-vs-histograms-in-project-management/]
<br>

#### Bean Plot
*A beanplot is an alternative to the boxplot for visual comparison of univariate data between groups. In a beanplot, the individual observations are shown as small lines in a one-dimensional scatter plot.* 
[https://cran.r-project.org/web/packages/beanplot/vignettes/beanplot.pdf] <br>

A bean plot can have one or multiple batches ("beans") displayed in the plot. The bean plot is a complete way to represent a distribution instead of the more summarised plots i.e. box plot.<br>
*It shows the smoothed density of points over a window called the 'bandwidth' so you can more easily represent more complex distributions accurately.*

<br>

#### Violin Plot
*A violin plot is more informative than a plain box plot. In fact while a box plot only shows summary statistics such as mean/median and interquartile ranges, the violin plot shows the full distribution of the data. The difference is particularly useful when the data distribution is multimodal (more than one peak).* [https://en.wikipedia.org/wiki/Violin_plot] <br> <BR>
Box plots show data points up to 1.5x the inter-quartile range i.e. the outliers, whereas violin plots display the entire data set.
<br>
    
![Violin](violin.PNG)

[https://matplotlib.org/gallery/statistics/boxplot_vs_violin.html]
<br>

http://yufree.github.io/blog/2013/08/15/boxplot-vs-violinplot.html <br>
It has been reported that violin plots appear useful for large sample sizes but with small sample sizes they may be misleading. With boxplots showing less data and more direct data, they seem to win as the cleaner choice for small samples.
https://figshare.com/articles/A_simple_comparison_of_box_plots_and_violin_plots/1544525
<br>

#### Barplots / Bar Chart
*Although bar charts are popular among researchers and are ubiquitous in quantitative software packages, they do not always provide the best visualization for a datset.*
[https://www.pmrjournal.org/article/S1934-1482(16)00067-8/pdf]

Box plots are simple alternative method showing variations providing more information. <br>
Barplots / Bar Charts are not the best option when it comes to summarising or comparing samples. A simple example of this links to a bar chart generally commencing at 0 and the bars travel vertically from this. This may not be an accurate representation when the dataset does not start at 0 or where the minimum value is a negative.
<br>

#### Bubble charts
*A bubble chart is a variation of a scatter chart in which the data points are replaced with bubbles, and an additional dimension of the data is represented in the size of the bubbles. Just like a scatter chart, a bubble chart does not use a category axis — both horizontal and vertical axes are value axes.*
https://support.office.com/en-us/article/present-your-data-in-a-bubble-chart-424d7bda-93e8-4983-9b51-c766f3e330d9
<br>
A bubble chart is a variation of a scatter plot where the data points are replaced by bubbles. Bubble charts are used when the data has three data series each with their own set of values. The sizing of each bubble is determined by the values in the third data series. <br>
Although a scatter plot / bubble chart will show more data, it is easier to compare data using a box plot. This is even more likely if you are dealing with a larger data set. 
<br>
![Scatter](scatter.PNG)

[https://www.itl.nist.gov/div898/handbook/pmc/section6/pmc612.htm]
<br>

#### Stem & Leaf Plot

An interesting comparison between Box Plots and Stem Plots was viewed on: http://www.ni.com/white-paper/3047/en/ <br>
*Stem-and-leaf displays are especially useful because they can show gaps in the data that are not obvious with the boxplot alone.* <br>
The box plot in the example gave the impresssion that the output was *symmetrical and evenly spaced*. Whereas the stem plot showed the data was not normally distributed, it showed two concentrations of data points with one at each end of the data range.<br>
The above information would lead me to believe that in order to have a true and clear view of the output, Stem Plot's seems to be a better option over the Box Plot.

## Findings

Box plots are a popular visual representation for comparing distributions between several groups or sets of data as they take up less space making it easier to visually analyse. [https://en.wikipedia.org/wiki/Box_plot]
<br>
When it comes to alternatives, Box Plots appear to be the superior choice when comparing distributions. 
*For example, comparing multiple histograms or stem-and-leaf plots is difficult because of the space they take.
Multiple density traces are difficult to compare when there are many of them plotted in one
plot, because the space becomes cluttered. Therefore, when comparing distributions between
batches, Tukey’s boxplot is commonly used.* [https://cran.r-project.org/web/packages/beanplot/vignettes/beanplot.pdf]
